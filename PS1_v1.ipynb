```python
#Vinita
# Problem Set 1
# Programming
# 2. Generalized Hailstone

def hailstone(x,a,b):
    sequence = [x]
#Calculating x from the general equation given
    while x >= 1:
        x = x//2 if x%2==0 else a*x+b
        sequence += [x]
# To check if there exist a convergence. Break the loop when the number of iterations goes beyond 100.
        if len(sequence) != len(set(sequence)):
            break
        if len(sequence) > 100:
            sequence = ['Stop']
            break
    return sequence
# Running a loop for a,b between 1 to 10 and x between 1 to 100
def func1():    
    for a in range(1,11):
        for b in range(1,11):
            sequence = []
            
            for x in range(1,101):
                sequence.append(hailstone(x,a,b))
            sequence = [x for x in sequence if x[0] != 'Stop']
           
    # Taking out the distinct patterns obtained for all the values of x for particular combination of a and b
            pattern = []
            for seq in sequence:
                var = sorted(seq)
                for i in range(len(var)-1):
                    if var[i] == var[i+1]:
                        y = var[i]
                pos = seq.index(y)
                pattern.append(seq[pos:])
            pattern = [list(x) for x in set(tuple(x) for x in pattern)]
            pattern.sort(key = len)
            if len(pattern) == 0:
                print("For",a,"&",b,", there are no holding patterns")
            else:
                print("For",a,"&",b,", the number of holding patterns are",len(pattern),". And those patterns are",pattern)

                              
%timeit func1()                
```
